```python
from os import listdir
from os.path import isfile, join
import librosa
from sklearn.manifold import TSNE
import numpy as np
import seaborn as sns
import pandas as pd
import openl3
import soundfile as sf
import os
```

```python
mypath = './slices2'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

hop_length = 512

X = np.empty([len(onlyfiles), 132])
```

```python
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

#for file in onlyfiles:
#    cmd = f'ffmpeg -i "slices2/{file}" "slices2/{file[:-4]}.wav"'
#    print(cmd)
#    os.system(cmd)
print(onlyfiles)
```

```python
audio = []
sr = []

i = 0

for file in onlyfiles:
    a, s = sf.read(f'slices2/{file}')
    audio.append(a)
    sr.append(s)

# Pass in a list of audio arrays and sample rates
emb_list, ts_list = openl3.get_audio_embedding(audio, sr, batch_size=32, hop_size=3, embedding_size=512)
```

```python
X = []
songs = []
for emb in emb_list:
    for vec in emb:
        X.append(vec)

X_embedded = TSNE(n_components=2, learning_rate=20, init='random').fit_transform(X)

i = 0
for emb in emb_list:
    song = []
    for vec in emb:
        song.append(X_embedded[i])
        i = i + 1
    songs.append(song)
        
print(songs[0][0][0])
    
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,10)

print(len(X_embedded))

x = X_embedded[:,0]
y = X_embedded[:,1]


for song in songs:
    x_new = []
    y_new = []
    
    for xy in song:
        x_new.append(xy[0])
        y_new.append(xy[1])
    
    plt.plot(x_new, y_new, '-', alpha=0.4)
    plt.plot(x_new, y_new, 'p')

plt.show()
```

```python
df = pd.DataFrame()
df["comp-1"] = X_embedded[:,0]
df["comp-2"] = X_embedded[:,1]

sns.scatterplot(x="comp-1", y="comp-2",
                palette=sns.color_palette("hls", 3),
                data=df).set(title="Iris data T-SNE projection") 
```
